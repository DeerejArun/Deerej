```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.tools.plotting import andrews_curves
from pandas.tools.plotting import parallel_coordinates
from pandas.tools.plotting import radviz
```

```python
Students = pd.read_csv("StudentsPerformance.csv")
Students.head()
```

```python
Students.info()
```

```python
sns.jointplot(x = "reading score", y = "writing score", data = Students, size = 5 )
```

```python
sns.FacetGrid(Students, hue = "gender", size = 5)\
   .map(plt.scatter, "reading score", "writing score")\
   .add_legend()
```

```python
sns.boxplot(x = "gender", y = "writing score", data = Students)
```

```python
ax = sns.boxplot(x = "gender", y = "reading score", data = Students)
ax = sns.stripplot(x = "gender", y = "reading score", data = Students, jitter = True, edgecolor = "gray")
```

```python
sns.violinplot(x = "gender", y = "writing score", data = Students, size = 5)
```

```python
sns.FacetGrid(Students, hue = "gender", size = 6)\
   .map(sns.kdeplot, "reading score")\
   .add_legend()
```

```python
sns.pairplot(Students.drop("gender", axis = 1), hue = "writing score", size = 5)
```

```python
sns.pairplot(Students.drop("gender", axis = 1), hue = "reading score", size = 3, diag_kind = "kde")
```

```python
Students.drop("reading score", axis = 1).boxplot(by = "gender", figsize = (12, 6))
```
